```python
def distribution_stats(data):
    stats = {
        "왜도(Skewness)": data.skew(),  # 양수면 오른쪽 치우침
        "첨도(Kurtosis)": data.kurtosis(),  # 정규분포 대비 뾰족한 정도
        "중앙값(Median)": data.median(),  # 평균보다 중앙값이 작으면 오른쪽 치우침
        "평균(Mean)": data.mean(),
        "평균/중앙값 비율": data.mean() / data.median()  # 1보다 크면 오른쪽 치우침
    }
    return pd.Series(stats)

```

```python
# 예시
stats = distribution_stats(data_df['Potential_Savings_Miscellaneous'])
print(stats)
```

```python
# 예시 Desired_Savings_Percentage
stats = distribution_stats(data_df['Income'])
print(stats)
```

```python
stats = distribution_stats(data_df['Desired_Savings_Percentage'])
print(stats)
```
